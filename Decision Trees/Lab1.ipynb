### Question 1.4
Create a solid line plot showing the average (across all nations)
life expectancy changing over time, with the standard deviation above and below shown with
filled between plot (use fill between plotting function with alpha = 0.3). Additionally, plot
the maximum life expectancy for each year and the minimum life expectancy for each year
as dotted line plots over time. Include legend for mean, minimum, and maximum, and donât
forget axes labels.


```python
minArr = []
maxArr = []
meanArr = []
stdArr = []
yr = np.unique(pop_df["Year"])
#for loop to get all the life expectancy's of each year to use various calculations
for i in range(2000,2016):
   LifeExpTotal = np.array(pop_df[pop_df["Year"] == i ]["Life_expectancy"].dropna())

   minArr.append(LifeExpTotal.min())
   maxArr.append(LifeExpTotal.max())
   meanArr.append(LifeExpTotal.mean())
   stdArr.append(LifeExpTotal.std())

   
plt.title("1.4 - Life Exp over 15 yrs")
plt.xlabel("Years")
plt.ylabel("Life Expectancy")
plt.plot(yr, minArr, label="Min")
plt.plot(yr, maxArr, label="Max")
plt.plot(yr, meanArr, label="Mean")
plt.fill_between(yr, np.array(meanArr) + np.array(stdArr), np.array(meanArr) - np.array(stdArr), alpha = 0.3)
plt.legend()
plt.show()
```

### Question 1.5

Extract Hepatitis B, Polio, Measles, BMI, Diphtheria, HIV/AIDS
for 2015 and remove the rows with NaN elements (use dropna()). Create the correlation
coefficient matrix (use np.corrcoef()) and create a heatmap showing the correlations, using
pcolor() and matplotlib. Label the rows and columns with the appropriate diseases. Include
a colorbar and a label on the colorbar. Which of these two are the most heavily correlated?

```python

features = ["Hepatitis_B", "Polio", "Diphtheria", "HIV/AIDS", "Measles"]
disease15 = np.array(pop_df[(pop_df["Year"] == 2015)][features].dropna(subset=features))

x = np.corrcoef(disease15, rowvar= False)

fig, ax = plt.subplots()
tmp = ax.pcolor(x, cmap = "PuRd" )

ax.set_xticks(np.arange(len(features)), labels=features)
ax.set_yticks(np.arange(len(features)), labels=features)

for i in range(len(features)):
      for j in range(len(features)):
         text = ax.text(j +0.5, i + 0.5, x[i,j], ha="center", va="center", color="b")

fig.colorbar(tmp).set_label("colorbar")
fig.tight_layout()
ax.set_title("Disease heat map")
plt.show()


```
